# Plot Gaussian experiment results
The notebook reads the CSV files generated by `run_gaussian.py` and plots the KL divergences.
It uses `matplotlib` and `pandas`.

```python
import pandas as pd
import matplotlib.pyplot as plt

results = []
for D in [4, 16, 64, 256]:
    df = pd.read_csv(f"results/gaussian_D{D}.csv")
    results.append(df)
    print(df)

plt.figure(figsize=(6,4))
for df in results:
    D = df.iloc[0,0]
    plt.subplot(2,2,D==4+1)
    plt.bar(['BaM', 'ADVI'], [df.iloc[0,3], df.iloc[0,4]])
    plt.title(f"D={D}")

plt.tight_layout()
plt.show()
```
